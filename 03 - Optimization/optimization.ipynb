# What is optimization?

Finding the values that optimize a given cost function within constraints

$ max\quad f_0(x) $

s.t.: 

$f_i(x) < 0, i=[1, ..., k] $

$h_j(x) = 0, j=[1, .., l]$


# Why should we care?

Machine learning problems can be formulated in the optimization framework. Where we have a cost function we try to optimize, in the case of machine learning it's usually the model parameters. We try to adjust the model parameters until getting the best fit to the data.

# Gradient descent

The function is kind of like a valley, we follow it down until we reach the bottom. The gradient gives us the direction of descent.

```python
x = np.linspace(-2, 2, 200)
y = np.linspace(-1, 2, 200)
xv, yv = np.meshgrid(x, y)

alpha=10
zv = (1-xv)**2 + alpha*(yv - xv**2)**2

x_values = []
y_values = []

x_hat = 0
y_hat = 1.5

for i in range(0, 1000):
    dx = -2*(1-x_hat) - 4*x_hat*alpha*(y_hat - x_hat**2)
    dy = 2*alpha*(y_hat- x_hat**2)

    x_hat -= 0.005*dx
    y_hat -= 0.005*dy
    
    x_values.append(x_hat)
    y_values.append(y_hat)


levels = np.arange(0, 10, 0.4)**2
plt.contour(xv, yv, zv, levels)  

plt.plot(x_values, y_values, '.')
```

```python
step = 0.01

m = 1
b = 0.5

x = np.arange(0, 1, step) 
error = np.random.rand(int(1/step))

y = m*x + b + 0.5*error 
```

```python
%matplotlib inline

plt.plot(x, y, '.')
plt.show()
```

$ y_{hat} = m_{hat}x + b_{hat} $

$e = \sum (y_{hat} - y)^2$ 

$\frac{de}{dm} = 2\sum \frac{y_{hat}}{dm}(y_{hat}-y) = 2\sum x \cdot (y_{hat}-y) $ 

$\frac{de}{db} = 2\sum \frac{y_{hat}}{db}(y_{hat}-y) = 2\sum (y_{hat}-y)$

```python
epochs = 1000

grad_step = 0.001

m_hat = np.random.rand(1)
b_hat = np.random.rand(1)

for i in range(1,epochs):
    dm = 2*np.sum(np.multiply(x,(m_hat*x + b_hat - y)))
    db = 2*np.sum(m_hat*x + b_hat - y)
    
    m_hat = m_hat - grad_step*dm
    b_hat = b_hat - grad_step*db
    

plt.plot(x,y,'.')
plt.plot(x, m_hat*x + b_hat)
plt.show()
```
